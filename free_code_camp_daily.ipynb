```python
def spookify(boo):
    # first to replace characters
    boo = boo.replace("_", "~").replace("-", "~")
    
    # determine the indices of the tilde characters
    tilde_idx = []
    for i in range(0, len(boo)):
        if boo[i] == '~':
            tilde_idx.append(i)
    
    #  modify the characters based on their position
    boo = list(boo.replace('~', ""))
    for i in range(0, len(boo)):
        if i % 2 == 0:
            boo[i] = boo[i].upper()
        else:
            boo[i] = boo[i].lower()
            
    # finally insert the tilde characters back into the modified string
    for i in tilde_idx:
        boo.insert(i, '~')
        
    boo = "".join(boo)
    return boo

spookify("hello_world-")
```

```python
# Prince's solution
def spookify(boo):
    boo = boo.strip().lower()
    spooky=""
    upper=False
    for i in boo:
        if i=="-" or i=="_":
            i="~"
        elif upper==False:
            i=i.upper()
            upper=True
        elif upper==True:
            upper=False
        spooky+=i
    
    boo=spooky.strip()

    return boo
```

```python
def nth_prime(n):
    num = 2
    n_prime = 2 # assume 2 is out last number for now
    num_nprime_found = 0 # tracks number of n primes found
    
    while True:
        prime_count = 0
        for i in range(2, num):
            if num % i == 0:
                prime_count += 1
        
        if prime_count != 0:
            num += 1
        else:
            num_nprime_found += 1
            n_prime = num
            num += 1
        
        if num_nprime_found == n:
            n = n_prime
            break
        
    return n
```
